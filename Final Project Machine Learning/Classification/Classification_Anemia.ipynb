### 2.1 Data dan Deskripsinya

Data yang digunakan dalam penelitian ini adalah data sekunder tentang jenis Anemia, yang diperoleh dari situs Kaggle [jenis Anemia](https://www.kaggle.com/datasets/ehababoelnaga/anemia-types-classification/data).
Data tersebut menggunakan parameter CBC (Complete Blood Count) yakni serangkaian pengukuran yang dilakukan pada sampel darah untuk memberikan informasi tentang berbagai komponen darah, termasuk sel darah merah, sel darah putih, dan trombosit. Parameter-parameter ini membantu dokter dalam mendiagnosis dan memantau berbagai kondisi kesehatan. Berikut 15 variabel yang digunakan.
1. HGB: Jumlah hemoglobin dalam darah, penting untuk pengangkutan oksigen.
2. PLT: Jumlah trombosit dalam darah, yang terlibat dalam pembekuan darah.
3. WBC: Jumlah sel darah putih, penting untuk respons imun.
4. RBC: Jumlah sel darah merah, yang bertanggung jawab untuk pengangkutan oksigen.
5. MCV (Mean Corpuscular Volume): Volume rata-rata satu sel darah merah.
6. MCH (Mean Corpuscular Hemoglobin): Jumlah rata-rata hemoglobin per sel darah merah.
7. MCHC (Mean Corpuscular Hemoglobin Concentration): Konsentrasi rata-rata hemoglobin dalam sel darah merah.
8. PDW: pengukuran variabilitas dalam distribusi ukuran trombosit dalam darah
9. PCT: Tes prokalsitonin dapat membantu penyedia layanan kesehatan Anda mendiagnosis apakah Anda mengalami sepsis akibat infeksi bakteri atau jika Anda berisiko tinggi mengalami sepsis
10. LYMp (Lymphocyte Percentage) : persentase limfosit (salah satu jenis sel darah putih) dari total jumlah sel darah putih dalam sampel darah.
11. NEUTp (Neutrophil Percentage) : Persentase neutrofil (jenis sel darah putih utama dalam melawan infeksi bakteri) dari total leukosit.
12. LYMn Lymphocyte Absolute Count (nilai absolut limfosit) : Jumlah absolut limfosit dalam satuan sel per mikroliter darah.
13. NEUTn Neutrophil Absolute Count (nilai absolut neutrofil) : Jumlah absolut neutrofil dalam darah.
14. HCT (Hematocrit) : persentase volume sel darah merah (eritrosit) dibandingkan dengan volume total darah dalam tubuh.
15. Diagnosis : Tiga jenis anemia Normocytic hypochromic anemia, Normocytic normochromic anemia dan Iron deficiency anemia.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/diagnosed_cbc_data_v4(d3).csv')
df.head()
```

#### **Preprocessing**

```python
print(df.info())
```

```python
# data dalam fitur Diagnosis
df['Diagnosis'].value_counts()
```

```python
# cek data duplikat
df.duplicated().sum()
```

```python
# cek nilai negatif
df[(df.drop(columns=['Diagnosis']) < 0).any(axis=1)]
```

```python
# hapus data duplikat dan nilai negatif
df_clean1 = df.drop_duplicates()
df_clean2 = df_clean1[(df_clean1.drop(columns=['Diagnosis']) >= 0).all(axis=1)]
df_clean2.info()
```

```python
counts = df_clean2["Diagnosis"].value_counts()
plt.figure(figsize=(4,4))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Diagnosis')
plt.axis('equal')
plt.show()
```

#### **Encoding**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_encode = df_clean2.copy()
df_encode['Diagnosis'] = le.fit_transform(df_clean2['Diagnosis'])
df_encode.head()
```

```python
df_encode['Diagnosis'].value_counts()
```

```python
num_columns = df_encode.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(12, 8))
for i, col in enumerate(num_columns, 1):
    plt.subplot(4, 4, i)
    sns.histplot(df_encode[col], kde=True, bins=30, color='red')
    plt.title(f'{col} Distribution')
    plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df_encode.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#### **Scaling**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = df_encode.copy()
df_scaled.iloc[:, :-1] = scaler.fit_transform(df_scaled.iloc[:, :-1])
df_scaled.head()
```

```python
# drop fitur Diagnosis sebagai target
x = df_scaled.drop(columns=['Diagnosis'])
y = df_scaled['Diagnosis']
print("x : ", x.shape)
print("y : ", y.shape)
```

#### **Split Data 80% Train 20% Test**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train,  y_test = train_test_split(x, y, test_size=0.2, random_state= 42)
```

```python
print(f"Jumlah data training: {len(x_train)}")
print(f"Jumlah data testing: {len(x_test)}")
```

### 2.2 Model  Logistic Regression
Logistic Regression adalah algoritma klasifikasi yang digunakan untuk memprediksi probabilitas dari suatu kategori atau kelas target. Model ini bekerja dengan mengubah kombinasi linier dari fitur (variabel input) melalui fungsi logistik (sigmoid) sehingga outputnya berada dalam rentang 0 hingga 1. Jika nilai probabilitas tersebut melewati ambang batas tertentu (biasanya 0.5), maka data diklasifikasikan ke dalam kelas tertentu.

Salah satu keunggulan Logistic Regression adalah kesederhanaannya dan kemudahannya untuk diinterpretasi. Setiap koefisien dalam model merepresentasikan seberapa besar pengaruh fitur tersebut terhadap kemungkinan suatu observasi termasuk ke dalam kelas positif. Karena itu, Logistic Regression sering digunakan dalam bidang medis dan sosial, di mana interpretasi model sangat penting. Selain itu, model ini juga relatif cepat dilatih dan tidak membutuhkan banyak sumber daya komputasi.

```python
# LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)

lr_y_train_pred = lr_model.predict(x_train)
lr_train_acc = accuracy_score(y_train, lr_y_train_pred)
print("Akurasi Train Logistic Regression: {:.2f}%".format(lr_train_acc*100))

lr_y_pred = lr_model.predict(x_test)
lr_acc = accuracy_score(y_test, lr_y_pred)
print("Akurasi Test Logistic Regression: {:.2f}%".format(lr_acc*100))

lr_report = classification_report(lr_y_pred, y_test)
```

### 2.3 Model K-Nearest Neighbors (KNN)
K-Nearest Neighbors (KNN) adalah algoritma klasifikasi berbasis instance-based learning atau lazy learning, yang berarti model ini tidak benar-benar melakukan pelatihan (training). Sebaliknya, KNN menyimpan seluruh data pelatihan, dan ketika ada data baru, model akan mengklasifikasikannya berdasarkan mayoritas kelas dari 'k' tetangga terdekat dalam ruang fitur. Jarak antara data biasanya dihitung menggunakan Euclidean Distance atau metrik jarak lainnya.

Keunggulan utama KNN adalah kesederhanaannya dan fleksibilitasnya. Algoritma ini tidak membuat asumsi tertentu tentang distribusi data, sehingga cocok digunakan pada berbagai jenis dataset, termasuk yang distribusinya tidak normal. KNN juga mampu menangani hubungan non-linier antara fitur dan label, selama jarak antar data representatif terhadap kelas. Namun, kelemahan dari KNN terletak pada efisiensi dan sensitivitas terhadap skala data. Ketika ukuran dataset besar, waktu prediksi bisa menjadi lambat karena setiap prediksi membutuhkan perhitungan jarak ke semua data latih.

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

k_values = range(1, 10)
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    knn_y_pred = knn.predict(x_test)
    accuracy_scores.append(accuracy_score(y_test, knn_y_pred))

plt.figure(figsize=(4, 3))
plt.plot(k_values, accuracy_scores, marker='o', linestyle='dashed')
plt.xlabel('Jumlah K')
plt.ylabel('Akurasi')
plt.title('Menentukan K Terbaik')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)

knn_y_train_pred = knn.predict(x_train)
knn_train_acc = accuracy_score(y_train, knn_y_train_pred)
print("Akurasi Train K-Nearest Neighbors: {:.2f}%".format(knn_train_acc*100))

knn_y_pred = knn.predict(x_test)
knn_acc = accuracy_score(knn_y_pred, y_test)
print("Akurasi Test K-Nearest Neighbors: {:.2f}%".format(knn_acc*100))

knn_report = classification_report(knn_y_pred, y_test)
```

### 2.4 Model Support Vector Machine (SVM)
Support Vector Machine (SVM) adalah algoritma klasifikasi yang bekerja dengan mencari hyperplane (garis batas) terbaik yang memisahkan kelas-kelas dalam ruang fitur. Tujuan dari SVM adalah memaksimalkan margin, yaitu jarak antara hyperplane dan titik data terdekat dari masing-masing kelas (support vectors). Semakin besar margin ini, semakin baik kemampuan model untuk menggeneralisasi data baru.

SVM sangat kuat untuk data berdimensi tinggi dan juga data yang tidak dapat dipisahkan secara linier dengan menggunakan teknik yang disebut kernel trick. Kernel ini memungkinkan data dipetakan ke ruang berdimensi lebih tinggi sehingga dapat ditemukan hyperplane pemisah. Kernel yang umum digunakan termasuk linear, polynomial, dan radial basis function (RBF). Karena kemampuannya ini, SVM banyak digunakan untuk klasifikasi kompleks seperti pengenalan wajah, teks, dan data medis. Meskipun akurat dan fleksibel, SVM memiliki beberapa kekurangan. Model ini kurang efisien pada dataset besar, karena waktu komputasi meningkat seiring bertambahnya jumlah data.

```python
# SVM
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(x_train, y_train)

svm_y_train_pred = svm_model.predict(x_train)
svm_train_acc = accuracy_score(y_train, svm_y_train_pred)
print("Akurasi Train Support Vector Machine: {:.2f}%".format(svm_train_acc*100))

svm_y_pred = svm_model.predict(x_test)
svm_acc = accuracy_score(y_test, svm_y_pred)
print("Akurasi Test Support Vector Machine: {:.2f}%".format(svm_acc*100))

svm_report = classification_report(svm_y_pred, y_test)
```

## 3. HASIL DAN PEMBAHASAN

```python
# Logistic Regression
print("Classification Report:")
print(lr_report)

lr_cm = confusion_matrix(y_test, lr_y_pred)
plt.figure(figsize=(4, 3))
sns.heatmap(lr_cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Logistic Regression Confusion Matrix')
plt.show()
```

```python
# K-Nearest Neighbors
print("Classification Report:")
print(knn_report)

knn_cm = confusion_matrix(y_test, knn_y_pred)
plt.figure(figsize=(4, 3))
sns.heatmap(knn_cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('K-Nearest Neighbors Confusion Matrix')
plt.show()
```

```python
# Support Vector Machine
print("Classification Report:")
print(svm_report)

svm_cm = confusion_matrix(y_test, svm_y_pred)
plt.figure(figsize=(4, 3))
sns.heatmap(svm_cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Support Vector Machine Confusion Matrix')
plt.show()
```

Berdasarkan hasil evaluasi, ketiga model klasifikasi yaitu Logistic Regression, K-Nearest Neighbors (KNN), dan Support Vector Machine (SVM) menunjukkan performa yang baik dalam mengklasifikasikan jenis anemia, namun dengan tingkat akurasi yang berbeda. Logistic Regression memiliku akurasi tertinggi sebesar 95%, diikuti oleh SVM dengan 93%, dan KNN dengan 89%. Logistic Regression memiliki nilai precision, recall, dan f1-score yang paling stabil di semua kelas, menjadikannya model paling akurat dan seimbang dalam eksperimen ini.

Sementara itu, model SVM memberikan hasil yang kompetitif, dengan nilai rata-rata metrik evaluasi sebesar 0,93. SVM sangat baik dalam mengklasifikasikan kelas 2 dan mampu menjaga keseimbangan antara precision dan recall pada semua kelas. Model ini juga menunjukkan jumlah kesalahan klasifikasi yang rendah, meskipun masih terdapat beberapa data kelas 0 dan 1 yang salah diprediksi. Di sisi lain, KNN menunjukkan performa paling rendah di antara ketiganya, dengan akurasi 89% dan kesalahan klasifikasi yang lebih tinggi terutama pada kelas 0 dan 1.

Meskipun demikian, setiap model memiliki karakteristik dan keunggulan masing-masing. Logistic Regression unggul dalam interpretasi dan efisiensi, SVM cocok untuk data kompleks dan berdimensi tinggi, sedangkan KNN dapat digunakan pada dataset kecil tanpa asumsi distribusi tertentu, namun sensitif terhadap skala dan outlier. Berdasarkan hasil pengujian, dapat disimpulkan bahwa Logistic Regression merupakan model terbaik dalam klasifikasi jenis anemia pada dataset ini, diikuti oleh SVM dan KNN.
